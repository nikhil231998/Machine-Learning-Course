## Task 3 : Data Cleaning & Preprocessing <a class="anchor" id="datacleaning"></a>

### Task 3.1 : Importing Packages <a class="anchor" id="import-package"></a>

All the necessary packages have been imported with the following code.

```python
#Importing packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
pd.set_option('display.max_columns', None)


plt.rcParams["figure.figsize"] = (12,7)
```

### Task 3.2 : Loading Dataset <a class="anchor" id="load-dataset"></a>

The original dataset consisted of two separate files for train and test. For the sake of this report, we have only taken the train.csv part. The number of rows after loading the data 100k,so we have taken subset of 5000 random rows from the combined data.

```python
#Reading the dataset
df = pd.read_csv('train.csv',encoding='unicode_escape')

#Sampling to select random 5000 rows
sampled_df = df.sample(5000, random_state = 999)

# Viewing the results
sampled_df.head()
```

### Task 3.3 : Removing Unnecessary Columns <a class="anchor" id="remove-col"></a>

Upon viewing the dataset, we see that we have 2 unnecessary columns, 'Unnamed: 0' and id are present in the dataset which are irrelevant to our analysis as they will lead to overfitting. In the below code, these columns are dropped.

```python
# Removing the 'Unnamed: 0' and 'id' columns
sampled_df.drop(['Unnamed: 0','id'], axis = 1, inplace = True)

# Viewing the results
sampled_df.head()
```

### Task 3.4 : Checking data types of the columns <a class="anchor" id="check-type"></a>

Once the data is loaded and unnecessary columns are removed, we will now check the data types of the columns and convert them into suitable format

```python
# Checking the column data types
sampled_df.dtypes
```

Upon viewing the data types, we see that the data type of columns starting from 'Inflight Wifi service' to 'Cleanliness' are of type integers, but should be ordinal categorical. Also, the data type of 'Arrival Delay in Minutes' column is float64, although we can work with it but for the sake of uniformity, we will change its data type to int64. The column 'Arrival Delay in Minutes' can not be straightaway converted to int64 because it has NA values, we will convert it to int64 after removing NA's in the next step.

```python
# Changing data types of columns from 'Inflight wifi service' to 'Cleanliness'
for i in range(6,20):
    sampled_df.iloc[:,i] = pd.Categorical(sampled_df.iloc[:,i],[0,1,2,3,4,5], ordered=True) 

# checking the result
sampled_df.dtypes
```

### Task 3.5 : Checking for missing values <a class="anchor" id="check-missing"></a>

In this step we will be finding the missing values and handling them

```python
#Checking for Missing Values
for i in range(len(list(sampled_df.columns))):
    print(f'{list(sampled_df.columns)[i]} = {sampled_df.iloc[:,i].isnull().sum()}')
```

We see that, there are 13 missing values in the 'Arrival Delay in Minutes' columns. Since the amount of missing values is minimal (less than 1%), therefore, can be ignored.

```python
# Deleing the NA values from the dataswet
sampled_df.dropna(axis =0, inplace = True)

# Checking the result
sampled_df['Arrival Delay in Minutes'].isnull().sum()
```

Now, we will convert the datatype of 'Arrival Delay in Minutes' column to int64.

```python
# Converting the datatype of'Arrival Delay in Minutes' column to int64
sampled_df['Arrival Delay in Minutes'] = sampled_df['Arrival Delay in Minutes'].astype('int64')
```

### Task 3.6 : Checking for typos & Whitespaces <a class="anchor" id="check-typo"></a>

In this step we will be looking for any typos and whitespaces in the non-numeric columns and rectifying them

```python
#Checking for Typos & Whitespace
for i in range(len(list(sampled_df.columns))):
    if sampled_df.iloc[:,i].dtypes == object:
        print(sampled_df.iloc[:,i].value_counts())
```

We see that there are no typos and whitespaces in the dataset.

### Task 4.2.4 : Services Provided vs Satisfaction <a class="anchor" id="two-var-plot-4"></a>

```python
def age_char_show(sampled_df):
    char_list = ['Inflight wifi service',
       'Departure/Arrival time convenient', 'Ease of Online booking',
       'Gate location', 'Food and drink', 'Online boarding', 'Seat comfort',
       'Inflight entertainment', 'On-board service', 'Leg room service',
       'Baggage handling', 'Checkin service', 'Inflight service',
       'Cleanliness']
    
    for col in char_list:
        sns.catplot(x=col,kind="count",data=sampled_df,orient="h",hue="satisfaction",legend=False)
        print("                  {} satisfaction".format(col))
        plt.subplots_adjust(bottom=0.2, left=.05, right=2.0, top=.90, hspace=500)
        plt.legend(loc='upper right')
        plt.xticks(rotation=90)
        plt.show()
        
age_char_show(sampled_df)
```

The above graph shows that services like On-board service, Food & Drink, Cleanliness, Inflight entertainment, Online boarding, Seat Comfort, Leg Room have direct impact on customer satisfaction whereas all the other services do not have greater impact.

### Task 4.3 : Three Variable Plots <a class="anchor" id="three-var-plot"></a>

### Task 4.3.1 : Effect of Gender & Class over Satisfaction <a class="anchor" id="three-var-plot-1"></a>

```python
from matplotlib.cm import get_cmap
from matplotlib.patches import Patch

#Taking subset of the three columns of the full dataset
sampled_dummy = sampled_df[['Class','Gender','satisfaction']]

# Make a dummy column, otherwise the groupby returns an empty df
sampled_dummy['Dummy'] = np.ones(len(sampled_dummy))

# This is now what I'd like to plot
grouped = sampled_dummy.groupby(by=['Class','Gender','satisfaction']).count().unstack()

# List of satisfaction, to use later as categories in subplots
kinds = grouped.columns.levels[1]

# colors for bar graph
colors = [get_cmap('viridis')(v) for v in np.linspace(0,1,len(kinds))]

sns.set(context="talk")
nxplots = len(grouped.index.levels[0])
nyplots = len(grouped.index.levels[1])
fig, axes = plt.subplots(nxplots,
                         nyplots,
                         sharey=True,
                         sharex=True,
                         figsize=(10,12))

fig.suptitle('Class,Gender and Satisfaction')

# plot the data
for a, b in enumerate(grouped.index.levels[0]):
    for i, j in enumerate(grouped.index.levels[1]):
        axes[a,i].bar(kinds,grouped.loc[b,j],color=colors)
        axes[a,i].xaxis.set_ticks([])

axeslabels = fig.add_subplot(111, frameon=False)
plt.tick_params(labelcolor='none', top=False, bottom=False, left=False, right=False)
plt.grid(False)
axeslabels.set_ylabel('Class',rotation='horizontal',y=1,weight="bold")
axeslabels.set_xlabel('Gender',weight="bold")

# x- and y-axis labels
for i, j in enumerate(grouped.index.levels[1]):
    axes[nyplots,i].set_xlabel(j)
for i, j in enumerate(grouped.index.levels[0]):
    axes[i,0].set_ylabel(j)

# Tune this manually to make room for the legend
fig.subplots_adjust(right=0.7)

fig.legend([Patch(facecolor = i) for i in colors],
           kinds,
           title="satisfaction",
           loc="center right")
```

We can observe that overall satisfaction patterns remains same for both genders when varied over class. Satisfaction of both gender who choose Business Class showed quite positive response & if the class is ECO or ECO-plus then chance of dissatisfaction increases drastically. In conclusion, we can say that satisfaction has quite high correlation with Class, & Gender would be a weak classifier of target variable(satisfaction).
